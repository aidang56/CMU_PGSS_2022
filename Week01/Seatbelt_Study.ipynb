```python
#Reading the file
import pandas as pd
seatbeltStudy = pd.read_csv("/gdrive/MyDrive/Comp. Sci. Lab Course/britishSeatBeltStudy.csv")

seatbeltStudy = seatbeltStudy.iloc[1:,:]

seatbeltStudy.columns = ['Drivers Killed', 'Drivers', 'Front', 'Rear', 'km/s', 'Gas Price', 'Van Killed', 'Law', 'Date']

seatbeltStudy['Law']  = seatbeltStudy['Law'].astype('category')

seatbeltStudy['Date'] = pd.to_datetime(seatbeltStudy['Date'], format='%Y/%m/%d')

seatbeltStudy
```

```python
seatbeltStudy.boxplot('Drivers Killed', 'Law')
```

```python
seatbeltStudy['Fraction of Drivers Killed'] = seatbeltStudy['Drivers Killed'] / seatbeltStudy['Drivers']
seatbeltStudy['Fraction of Drivers Killed'].describe
seatbeltStudy
```

```python
seatbeltStudy.boxplot('Fraction of Drivers Killed', 'Law')
```

```python
#BOXPLOT
seatbeltStudy.groupby('Law').agg({'Drivers Killed': ['min', 'max', 'mean', 'std'], 'Fraction of Drivers Killed': ['min', 'max', 'mean', 'std'], 
                                  'Gas Price': ['min', 'max', 'mean', 'std']})
```

```python
seatbeltStudy['Fraction of Drivers Killed in the Front Seat'] = seatbeltStudy['Drivers Killed'] / seatbeltStudy['Front']
seatbeltStudy['Fraction of Drivers Killed in the Front Seat'].describe
seatbeltStudy
```

```python
seatbeltStudy.boxplot('Fraction of Drivers Killed in the Front Seat', 'Law')
```

```python
seatbeltStudy.plot.scatter(x = 'Gas Price', y = 'km/s', c = 'Law', colormap = 'viridis')
```

```python
seatbeltStudy.plot.scatter(x = 'Date', y = 'Fraction of Drivers Killed', c = 'Law', colormap = 'viridis')
```

```python
seatbeltStudy.plot.scatter(x = 'Drivers', y = 'Drivers Killed', c = 'Law', colormap = 'viridis')
```
