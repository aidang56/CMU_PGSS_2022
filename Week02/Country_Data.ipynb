#Country Data Exploration

```python
#Reading the file
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

countryData = pd.read_csv("/content/drive/MyDrive/Comp. Sci. Lab Course/factbook.csv")
countryData = countryData.iloc[1:,:]

countryData['Electricity - consumption(kWh)']  = countryData['Electricity - consumption(kWh)'].astype('float')
countryData['GDP']  = countryData['GDP'].astype('float')
countryData['Area(sq km)']  = countryData['Area(sq km)'].astype('float')
countryData['Current account balance']  = countryData['Current account balance'].astype('float')
countryData['Birth rate(births/1000 population)']  = countryData['Birth rate(births/1000 population)'].astype('float')
countryData['Death rate(deaths/1000 population)']  = countryData['Death rate(deaths/1000 population)'].astype('float')
countryData['Debt - external']  = countryData['Debt - external'].astype('float')
countryData['Electricity - production(kWh)']  = countryData['Electricity - production(kWh)'].astype('float')
countryData['Exports']  = countryData['Exports'].astype('float')
countryData['Oil - production(bbl/day)']  = countryData['Oil - production(bbl/day)'].astype('float')
countryData['Oil - proved reserves(bbl)']  = countryData['Oil - proved reserves(bbl)'].astype('float')
countryData['Population']  = countryData['Population'].astype('float')
countryData['Public debt(% of GDP)']  = countryData['Public debt(% of GDP)'].astype('float')
countryData['Railways(km)']  = countryData['Railways(km)'].astype('float')
countryData['Reserves of foreign exchange & gold']  = countryData['Reserves of foreign exchange & gold'].astype('float')
countryData['Telephones - main lines in use']  = countryData['Telephones - main lines in use'].astype('float')
countryData['Telephones - mobile cellular']  = countryData['Telephones - mobile cellular'].astype('float')
countryData['Total fertility rate(children born/woman)']  = countryData['Total fertility rate(children born/woman)'].astype('float')
countryData['Unemployment rate(%)']  = countryData['Unemployment rate(%)'].astype('float')


countryData
```

```python
countryDataOutlierExcluded = countryData[countryData['GDP'] < 1.01e+12]
countryDataOutlierExcluded = countryDataOutlierExcluded[countryDataOutlierExcluded['Population'] < 5e11]
countryDataOutlierExcluded = countryDataOutlierExcluded[countryDataOutlierExcluded['Railways(km)'] < 1.01e+10]
```

```python
countryDataOutlierExcluded.plot.scatter(x = 'GDP', y = 'Electricity - consumption(kWh)', c = 'Area(sq km)', colormap = 'viridis')
```

```python
countryDataOutlierExcluded.plot.scatter(x = 'GDP', y = 'Public debt(% of GDP)', c = 'Population', colormap = 'viridis')
```

```python
countryData.plot.scatter(x = 'Birth rate(births/1000 population)', y = 'Death rate(deaths/1000 population)', c = 'Population', colormap = 'viridis')
```

```python
countryDataOutlierExcluded.plot.scatter(x = 'Railways(km)', y = 'Electricity - consumption(kWh)', c = 'Population', colormap = 'viridis')
```

```python
countryDataOutlierExcluded.plot.scatter(x = 'Unemployment rate(%)', y = 'Public debt(% of GDP)', c = 'Population', colormap = 'viridis')
```
